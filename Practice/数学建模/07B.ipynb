```python
label_dir = 'C:/Users/Forry/Desktop/第二期培训/历年赛题/2007/公交路线/1.1 公汽线路信息.txt'

label_folder = open(label_dir, "r")#,encoding='utf-8'
trainlines = label_folder.read().splitlines() #返回每一行的数据
i=0
#print(trainlines)
#print(len(trainlines))
k=[]
for line in trainlines:
    #print(line)
    a = list(line)
    #print(a)
    if a==[]:
        #print(1)
        continue

    elif a[0]== 'L':
        s=[]
        s.append(line)
        #print(s)
    elif a[0]== '分' :
        s.append(line)
       # print(s)
        #print(1)
    elif a[0]== '单' :
        s.append(line)
       # print(s)
        #print(1)    
    elif a[0]== '上':
        #print(1)
        b= line.split("-") #按照空格键分割每一行里面的数据
        #print(b)
        del b[0]
        s.extend(b)
        #print(s)
        k.append(s)
    elif a[0]== '环':
        #print(1)
        b= line.split("-") #按照空格键分割每一行里面的数据
        #print(b)
        del b[0]
        s.extend(b)
        #print(s)
        k.append(s)
    elif a[0]=='S':
        #print(1)
        b= line.split("-") #按照空格键分割每一行里面的数据
        s.extend(b)
        #print(s)
        k.append(s)
    else:
        continue

    #i=i+1
print(k) 
#train_box = load_train_test_set(test_txt)
```

```python
import pandas as pd

s = pd.DataFrame(k)
f = 'C:/Users/Forry/Desktop/2.xlsx'
s.to_excel(f)#输出至源文件当前路径
```
