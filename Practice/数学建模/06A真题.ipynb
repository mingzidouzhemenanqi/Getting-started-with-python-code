```python
#九个分社旧书使用率
jiuk = d2001['Q2c学科类别(Q2c)'] 
g = jiuk.value_counts()
#获取九个分社
fix=list(g.index)
#print(fix)

y = 2001
for i in [d2001, d2002, d2003, d2004,d2005]:
    #获取各出版社书量
    jiuk = i['Q2c学科类别(Q2c)'] 
    o = jiuk.value_counts()

    a = i[['Q2k是否为旧书（1.否；2.是）(Q2k)','Q2c学科类别(Q2c)']]
    f = a.loc[(a["Q2k是否为旧书（1.否；2.是）(Q2k)"]==1)]
    p=0
    for j in fix:
        k = f.loc[(f["Q2c学科类别(Q2c)"]==j)]
        #计算某分社旧书使用率
        c = len(k)
        d = int(o[j])
        p = p+1
        print("{}年{}分社旧书使用率为:{}".format(y,j,(c/d)))
    y = y+1
        
        
    
    
```
